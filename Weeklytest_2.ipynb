```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'
}
```

```python
url = "https://stackoverflow.com/questions/415511/how-do-i-get-the-current-time-in-python"
```

```python
response = requests.get(url,headers=headers)
```

```python
html_content = response.content
print(html_content)
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
question = soup.find('a', class_='question-hyperlink').text
```

```python
answers = soup.find_all('div', class_='answer')
```

```python
answers_text = []
for answer in answers:
    answer_text = answer.find('div', class_='s-prose js-post-body').text
    answers_text.append(answer_text)
```

```python
answers_text
```

```python
print("Question: ", question)
print("\nAnswers:")
for i, answer in enumerate(answers_text, 1):
    print(f"Answer {i}: {answer}\n")
```
